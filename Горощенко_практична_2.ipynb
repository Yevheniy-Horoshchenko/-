```python
# Завдання 10
# Знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату
import pandas as pd

# Зчитуємо дані з вказаного URL
url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

data = pd.read_html(url, header=0)[2]  # Обираємо другий елемент списку, де знаходиться таблиця

data.replace(to_replace=r'^-$', value=pd.NA, regex=True, inplace=True)

data = data.apply(pd.to_numeric, errors='ignore')

max_salary_year = data['Рік'][data['В національній валюті'].idxmax()]
max_salary_value = data['В національній валюті'].max()

print(f"Рік з найвищою зарплатою: {max_salary_year}")
print(f"Найвища зарплата в гривнях: {max_salary_value}")
```

```python
# Завдання 12
# Виведіть роки, коли зарплата в грн була вище середньої 
import pandas as pd

# Зчитуємо дані з вказаного URL
url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

data = pd.read_html(url, header=0)[2]  # Обираємо другий елемент списку, де знаходиться таблиця

data.replace(to_replace=r'^-$', value=pd.NA, regex=True, inplace=True)

data['В національній валюті'] = pd.to_numeric(data['В національній валюті'].replace(['▲', '▼'], '', regex=True), errors='coerce')

average_salary_uah = data['В національній валюті'].mean()

above_average_years = data['Рік'][data['В національній валюті'] > average_salary_uah]
print("Роки, коли зарплата в гривнях була вище середньої:")
print(above_average_years)
```

```python
# Завдання 16
# Обчисліть зарплату за останні два роки і порівняйте її із зарплатою за останні 5 років. Вкажіть, чи є різниця, і в якому напрямку. 
import pandas as pd

# Зчитуємо дані з вказаного URL
url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

data = pd.read_html(url, header=0)[2]  # Обираємо другий елемент списку, де знаходиться таблиця


data['В національній валюті'] = pd.to_numeric(data['В національній валюті'].str.replace('[^0-9..]', '', regex=True), errors='coerce')

data = data.iloc[:-1]

average_salary_last_two_years = data['В національній валюті'].tail(2).mean()
print('Зарплата за останні два роки:')
print(average_salary_last_two_years)

average_salary_last_five_years = data['В національній валюті'].tail(5).mean()
print('Зарплата за останні 5 рокiв:')
print(average_salary_last_five_years)

if average_salary_last_two_years > average_salary_last_five_years:
    result = "Зарплата за останні два роки вища за середню за останні 5 років."
elif average_salary_last_two_years < average_salary_last_five_years:
    result = "Зарплата за останні два роки нижча за середню за останні 5 років."
else:
    result = "Зарплата за останні два роки дорівнює середній зарплаті за останні 5 років."

print(result)
```
