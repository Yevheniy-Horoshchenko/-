```python
# Завдання 1
# Прочитайте csv файл (метод read_csv)
import pandas as pd

file_path = 'C:/Users/38099/Desktop/archive (1)/bestsellers with categories.csv'

data = pd.read_csv(file_path, encoding='cp1251')
print(data)
```

```python
# Завдання 2
# Виведіть перші 10 рядків (метод head) 
data.head(10)
```

```python
# Завдання 3
# Виведіть розміри датасету (метод shape) 
data.shape
```

```python
# Завдання 4
# Змініть назви всіх колонок: назва з маленької букви, без пробілів через підкреслення. 

data.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
data.head(10) 
```

```python
# Завдання 5
# Вивести кількість пропусків у кожному стовпчику
missing_values = data.isna().sum()

print("Кількість пропусків у кожному стовпці:")
print(missing_values)
```

```python
# Завдання 6
# За допомогою методу unique  перевірте. Які є унікальні жанри в стовпці genre  
unique_genres = data['genre'].unique()
print("Унікальні жанри в стовпці 'genre':")
print(unique_genres)
```

```python
# Завдання 7
# Визначте максимальну, мінімальну, медіанну ціну за допомогою методів  max, min, mean, median 
max_price = data['price'].max()
min_price = data['price'].min()
mean_price = data['price'].mean()
median_price = data['price'].median()

print("Максимальна ціна:", max_price)
print("Мінімальна ціна:", min_price)
print("Середня ціна:", mean_price)
print("Медіанна ціна:", median_price)
```

```python
# Завдання 8 
# Виконайте наступні завдання з пошуку і сортування даних 
# Який рейтинг у датасеті найвищий
max_rating = data['user_rating'].max()

print("Найвищий рейтинг:", max_rating)
print('\n')
# Скільки книг мають такий рейтинг?
count_books_with_rating_5 = data[data['user_rating'] == max_rating].shape[0]

print("Кількість книг з рейтингом 4.9:", count_books_with_rating_5)
print('\n')
# Яка книга має найбільше відгуків? 
index_of_book_with_max_reviews = data['reviews'].idxmax()

book_with_max_reviews = data.loc[index_of_book_with_max_reviews]

print("Книга з найбільшою кількістю відгуків:")
print(book_with_max_reviews['name'])
print('\n')
# З тих книг, що потрапили до Топ-50 у 2010 році, яка книга найдорожча ?
top_50_2010 = data[(data['year'] == 2010) & (data['user_rating'] <= 50)]

most_expensive = top_50_2010.loc[top_50_2010['price'].idxmax()]

print("Найдорожча книга серед тих, що потрапили до Топ-50 у 2010 році:")
print(most_expensive['name'])
print('\n')
# Скільки книг жанру Fiction потрапили до Топ-50 у 2012 році
top_50_2012 = data[(data['year'] == 2012) & (data['user_rating'] <= 50)]

fiction_books = top_50_2012[top_50_2012['genre'] == 'Fiction']

count_fiction_books_in_2012 = fiction_books.shape[0]

print("Кількість книг жанру Fiction у Топ-50 у 2012 році:", count_fiction_books_in_2012)
print('\n')
# Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або метод isin)? 
target_years = [2010, 2011]
target_rating = 4.9

filtered_books = data[((data['year'] == 2010) | (data['year'] == 2011)) & (data['user_rating'] == target_rating)]

count_books = filtered_books.shape[0]

print(f"Кількість книг з рейтингом 4.9 у 2010, 2011 роках: {count_books}")
print('\n')
```

```python
# Завдання 9
# Вивести максимальну і  мінімальну ціну для жанру Fiction I NonFiction (за допомогою методів  groupby та agg, для підрахунку мінімальних та максимальних значень використовуйте max та min)
result = data.groupby('genre').agg({'price': ['min', 'max']})

print(result)
```
