```python
# Завдання 1
# Створіть одновимірний масив (вектор) зі зворотнім порядком перших 10 натуральних чисел та виведіть його значення. 
reverse_list = list(range(10, 0, -1))

print(reverse_list)
```

```python
# Завдання 2
# Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його послідовністю чисел, де кожне наступне число менше попереднього на 3, та виведіть його значення
rows = 3
cols = 3

matrix = [[0] * cols for _ in range(rows)]

current_value = 9

for i in range(rows):
    for j in range(cols):
        matrix[i][j] = current_value
        current_value -= 3

print("Двовимірний масив (матриця):")
for row in matrix:
    print(row)
```

```python
# Завдання 4
# Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та обчисліть добуток всіх елементів. 
import numpy as np

rows = 3
cols = 3

matrix = np.random.rand(rows, cols)

print("Двовимірний масив (матриця):")
print(matrix)

product = np.prod(matrix)

print("Добуток всіх елементів матриці:", product)
```

```python
# Завдання 5
# Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від -20 до 20 та виконайте на них поелементні операції додавання, віднімання та множення. 
import numpy as np

size = 6

array1 = np.random.randint(-20, 21, size)
array2 = np.random.randint(-20, 21, size)

print("Перший масив:", array1)
print("Другий масив:", array2)

add_result = np.add(array1, array2)
print("Результат додавання:", add_result)

subtraction_result = np.subtract(array1, array2)
print("Результат віднімання:", subtraction_result)

multiplication_result = np.multiply(array1, array2)
print("Результат множення:", multiplication_result)
```

```python
# Завдання 6
# Створіть два вектори розміром 8, заповніть їх довільними цілими числами та знайдіть їх скалярний добуток. 
import numpy as np

size = 8

vector1 = np.random.randint(-10, 11, size)
vector2 = np.random.randint(-10, 11, size)

print("Перший вектор:", vector1)
print("Другий вектор:", vector2)

dot_product = np.dot(vector1, vector2)
print("Скалярний добуток:", dot_product)
```

```python
# Завдання 7
# Створіть дві матриці розміром 3x2 та 2x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 8 та перемножте їх між собою. 
import numpy as np

rows1 = 3
cols1 = 2 
rows2 = 2
cols2 = 2

matrix1 = np.random.randint(1, 9, size=(rows1, cols1))
matrix2 = np.random.randint(1, 9, size=(rows2, cols2))

print("Перша матриця:")
print(matrix1)
print("\nДруга матриця:")
print(matrix2)

result_matrix = np.dot(matrix1, matrix2)

print("\nРезультат перемноження матриць:")
print(result_matrix)
```

```python
# Завдання 8
# Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть її обернену матрицю
import numpy as np

rows, cols = 2, 2

matrix = np.random.randint(1, 11, size=(rows, cols))

print("Початкова матриця:")
print(matrix)

inverse_matrix = np.linalg.inv(matrix)
print("\nОбернена матриця:")
print(inverse_matrix)
```

```python
# Завдання 9
# Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 0 до 1 та транспонуйте її. 
import numpy as np

rows, cols = 3, 3

matrix = np.random.randint(0, 2, size=(rows, cols))

print("Початкова матриця:")
print(matrix)

transposed_matrix = np.transpose(matrix)

print("\nТранспонована матриця:")
print(transposed_matrix)
```

```python
# Завдання 10 
# Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть добуток матриці на вектор. 
import numpy as np

rows, cols = 3, 4
vector_size = 4

matrix = np.random.rand(rows, cols)

vector = np.random.randint(1, 11, size=vector_size)

print("Матриця:")
print(matrix)
print("\nВектор:")
print(vector)

product_result = np.dot(matrix, vector)

print("\nРезультат добутку матриці на вектор:")
print(product_result)
```

```python
# Завдання 11
# Створіть матрицю розміром 2x3 та вектор розміром 3, заповніть матрицю випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть суму добутків елементів кожного рядка матриці на відповідний елемент вектору. 
import numpy as np

rows, cols = 2, 3
vector_size = 3

matrix = np.random.randint(1, 11, size=(rows, cols))

vector = np.random.randint(1, 11, size=vector_size)

print("Матриця:")
print(matrix)
print("\nВектор:")
print(vector)

result_sum = np.sum(matrix * vector, axis=1)

print("\nРезультат суми добутків елементів кожного рядка матриці на відповідний елемент вектору:")
print(result_sum)
```

```python
# Завдання 15
# Створіть матрицю розміром 7x7, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить найбільший елемент у кожному рядку матриці. 
import numpy as np

rows, cols = 7, 7

matrix = np.random.rand(rows, cols)

print("Матриця:")
print(matrix)

max_elements_vector = np.max(matrix, axis=1, keepdims=True)

print("\nВектор-стовпчик з найбільшими елементами у кожному рядку:")
print(max_elements_vector)
```
